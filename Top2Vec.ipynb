```python
import top2vec
from top2vec import Top2Vec


def top2vec_fit(df):
  # Turn the corresponding column to the right form
    documents = df["review"].tolist()

  # Train the Top2vec model
    top2vec_model = Top2Vec(documents=documents, embedding_model="universal-sentence-encoder")

    return top2vec_model
```

```python
def print_topics(model):
  # The number of topics that Top2Vec has found in the data
    print("The number of topics is: ", model.get_num_topics())

  # Get the first 20 topics in decreasing size.
  # topic_words: For each topic the top 50 words are returned, in order of semantic similarity to topic.
  # word_scores: For each topic the cosine similarity scores of the top 50 words to the topic are returned.
  # topic_nums: The unique index of every topic will be returned
    topic_words, word_scores, topic_nums1 = model.get_topics(20)
    zipped_data = zip(topic_nums1, topic_words, word_scores)

  # topic_sizes: The number of documents most similar to each topic.
    topic_sizes, topic_nums2 = model.get_topic_sizes()


  # Print words in these topics

    df1 = pd.DataFrame(zipped_data, columns=['Topic', 'Representation', 'Scores'])
    df2 = pd.DataFrame({'Topic': topic_nums2, 'Count': topic_sizes})
    merged_df = pd.merge(df2, df1, on='Topic', how='inner')
    return merged_df
```

# PLAYERUNKNOWN'S BATTLEGROUNDS

```python
top2vec_model_1 = top2vec_fit(df_player)
```

```python
result1 = print_topics(top2vec_model_1)
result1
```

# Grand Theft Auto V

```python
top2vec_model_2 = top2vec_fit(df_grand)
```

```python
result2 = print_topics(top2vec_model_2)
result2
```
